reshape() — Changing shape of array


```python
import numpy as np
a = np.array([1, 2, 3, 4, 5, 6, 5,34])
reshaped = a.reshape(4,2)
print(reshaped)
```

flatten() — Make multidimensional to 1D

```python
arr = np.array([[1, 2], [3, 4]])
flat = arr.flatten()
print(flat)
```

axis in operations (important!)

```python
a = np.array([[1, 2], [3, 4]])
print(np.sum(a, axis=0))  # sum column-wise: [4, 6]
print(np.sum(a, axis=1))  # sum row-wise: [3, 7]
```

np.where() — condition-based output

```python
a = np.array([10, 15, 20, 25])
result = np.where(a > 18, 'High', 'Low')
print(result)
```

np.unique() — Find unique values

```python
a = np.array([1, 2, 2, 3, 3, 3,7,54,3,3,545,4,6,4577,5,7])
print(np.unique(a))
```

np.concatenate() — Join arrays

```python
a = np.array([1, 2])
b = np.array([3, 4])
joined = np.concatenate((a, b))
print(joined)
```

np.random Module (for ML use)

```python
print(np.random.randint(1, 10, size=(2, 3)))  # random integers 1–9
print(np.random.rand(2, 2))                  # random floats 0–1
```

np.linspace() — evenly spaced numbers

```python
print(np.linspace(0, 1, 5))  # 5 numbers between 0 and 1
```

Boolean Indexing (filtering)

```python
a = np.array([10, 20, 30, 40])
print(a[a > 20])  # filter elements greater than 20
```

np.argmax() & np.argmin() — index of max/min



```python
a = np.array([10, 50, 30])
print(np.argmax(a))  # index of max (1)
print(np.argmin(a))  # index of min (0)
```
