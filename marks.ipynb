```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create the dataset
data = {
    'Name': ['Lakshana', 'Keerthana', 'Harshini', 'Lavanthika', 'Hemala'],
    'Maths': [95, 88, 76, 68, 90],
    'Physics': [92, 85, 80, 70, 88],
    'Chemistry': [78, 67, 85, 65, 92],
    'Biology': [85, 75, 72, 60, 89],
    'English': [91, 82, 78, 72, 86]
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)

# Step 2: Calculate total and average marks
df['Total'] = df[['Maths', 'Physics', 'Chemistry', 'Biology', 'English']].sum(axis=1)
df['Average'] = df['Total'] / 5
print("\nData with Total and Average:")
print(df)

# Step 3: Identify topper(s)
topper_total = df['Total'].max()
topper_row = df[df['Total'] == topper_total]
print("\nTopper:")
print(topper_row[['Name', 'Total', 'Average']])

# Step 4: Count students with average > 75
above_75 = df[df['Average'] > 75]
print("\nStudents with Average > 75:")
print(above_75[['Name', 'Average']])

plt.figure(figsize=(8,5))
plt.bar(df['Name'], df['Total'], color='skyblue')
plt.title('Total Marks of Students')
plt.xlabel('Name')
plt.ylabel('Total Marks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

topper_index = topper_row.index[0]
subject_scores = df.loc[topper_index, ['Maths', 'Physics', 'Chemistry', 'Biology', 'English']]
plt.figure(figsize=(6,6))
plt.pie(subject_scores, labels=subject_scores.index, autopct='%1.1f%%', startangle=90)
plt.title(f"{df.loc[topper_index, 'Name']}'s Subject-wise Score Distribution")
plt.show()

# Step 7: Line chart – Average score trend
plt.figure(figsize=(8,5))
plt.plot(df['Name'], df['Average'], marker='o', linestyle='-', color='green')
plt.title('Average Marks Trend')
plt.xlabel('Name')
plt.ylabel('Average Marks')
plt.grid(True)
plt.tight_layout()
plt.show()
```
