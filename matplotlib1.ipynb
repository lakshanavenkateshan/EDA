```python
x=[0,1,3,4,5,6]
y=[0,4,5,3,2,6]
plt.plot(x,y)
plt.title("my plot", fontdict={'fontname': 'Liberation Sans', 'fontsize': 13})
plt.xlabel("x",fontdict={'fontname': 'Liberation Sans','fontsize': 13 })
plt.ylabel("y",fontdict={'fontname': 'Liberation Sans','fontsize': 13})
plt.xticks([0,2,4,6,8,10])
plt.yticks([0,2,4,6,8,10])
plt.show()
```

```python
x=[0,1,3,4,5,6]
y=[0,4,5,3,2,6]
plt.plot(x,y, label='2x',color='red',linewidth=1,linestyle='--',marker='.',markersize=10,markeredgecolor='blue')
plt.title("my plot", fontdict={'fontname': 'Liberation Sans', 'fontsize': 13})
plt.xlabel("x",fontdict={'fontname': 'Liberation Sans','fontsize': 13 })
plt.ylabel("y",fontdict={'fontname': 'Liberation Sans','fontsize': 13})
plt.xticks([0,2,4,6,8,10])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
x=[0,1,3,4,5,6]
y=[0,4,5,3,2,6]
#plt.plot(x,y, label='2x',color='red',linewidth=1,linestyle='--',marker='.',markersize=10,markeredgecolor='blue')
#use shorthand notation
plt.plot(x,y,'go-',label='2x')
#line number two
x2=np.arange(0.5,0.6)
plt.plot(x2,x2**2,'r',label='X^2')
#valid strings:
#'b^--'	blue color, triangle marker, dashed line
#'r.--'	red color, dot marker, dashed line
#'go-'	green color, circle marker, solid line
plt.title("my plot", fontdict={'fontname': 'Liberation Sans', 'fontsize': 13})
plt.xlabel("x",fontdict={'fontname': 'Liberation Sans','fontsize': 13 })
plt.ylabel("y",fontdict={'fontname': 'Liberation Sans','fontsize': 13})
plt.xticks([0,2,4,6,8,10])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.figure(figsize=(3,3),dpi=160)
plt.plot(x,y, 'b^--',label='2x')
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:5],x2[:5]**2,'r',label='x^2')
plt.plot(x2[4:],x2[4:]**2,'r--')
plt.title("my plot", fontdict={'fontname': 'Liberation Sans', 'fontsize': 13})
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4,])
plt.legend()
plt.savefig('mygraph.png',dpi=300)
plt.show()
```

BAR CHART


```python
labels=['A','B','C']
values=[2,6,3]
bars=plt.bar(labels,values)
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')
#in for loop cases
patterns=['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.figure(figsize=(2,2),dpi=100)
plt.show()
```
