```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 15, 7]

plt.bar(x, y)
plt.title("Bar Chart")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [5, 7, 6, 8]

plt.scatter(x, y)
#plt.plot(x, y, color='red')
#plt.plot(x, y, linestyle='--')
#plt.plot(x, y, marker='o')
#plt.grid(True)
#plt.legend(['Line 1'])
plt.title("Scatter Plot")
plt.show()
```

```python
labels = ['Python', 'C', 'Java']
sizes = [50, 25, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
plt.subplot(1, 2, 1)  # 1 row, 2 cols, 1st plot
plt.plot([1, 2, 3], [4, 5, 6])
plt.subplot(1, 2, 2)  # 2nd plot
plt.bar(['A', 'B'], [10, 20])
plt.show()
```

Histograms — useful in ML to visualize distributions

```python
data = [1,2,2,3,3,3,4,4,4,4]
plt.hist(data)
plt.show()
```
