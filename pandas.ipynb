groupby() — Group and Analyze Data

```python
import pandas as pd
data={
    'country':['south korea','japan','america','germany','brazil','france'],
    'budget':[120000,130000,200000,100000,210000,230000]
}
df=pd.DataFrame(data)
grouped=df.groupby('country')['budget'].mean()
grouped=df.groupby('country')['budget'].sum()
grouped=df.groupby('country')['budget'].count()

print(grouped)
```

apply() + lambda — Apply Custom Logic


```python
import pandas as pd
data={
    'job' : ['ceo','baker','manager','bookshop'],
    'salary' : [70000,30000,40000,12000]
}
df=pd.DataFrame(data)
df['final salary'] = df['salary'].apply(lambda x: x+x*0.5)
#print(df)
df['result']=df['final salary'].apply(lambda x: 'vip' if x>=100000 else('good' if 60000<=x<=90000 else 'not enough'))
print(df)
```

Sorting — sort_values() + sort_index()


```python
import pandas as pd

data = {
    'salary': [680000, 24000, 36000, 4000, 95000],
    'bonus': [7000, 2000, 6000, 9000, 5000]
}

df = pd.DataFrame(data)

print(df.sort_values(by='salary', ascending=False))
print(df.sort_index())
```

Merging


```python
import pandas as pd
df1 =({
    'id':[1,2,3,4,5],
    'name':['lakshana','keerthana','hemala','kamali','harshini']
})
df2=({
    'id':[1,2,5,8,4],
    'score':[23,34,5,2,1]
})
df1=pd.DataFrame(df1)
df2=pd.DataFrame(df2)
result=pd.merge(df1,df2,on='id',how='inner')
print(result)
```

join

```python

df1 = pd.DataFrame({'name': ['Ava', 'Ben', 'Cleo']}, index=[1, 2, 3])
df2 = pd.DataFrame({'score': [90, 85, 78]}, index=[1, 2, 4])

result = df1.join(df2, how='inner')
print(result)
```
