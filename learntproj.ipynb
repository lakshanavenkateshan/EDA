```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    'name':['lakshana','keerthana','harshini','lavanthika'],
    'maths':[94,98,76,64],
    'physics':[97,87,94,92],
    'chemistry':[65,43,78,98]
}
df=pd.DataFrame(data)
print(df)
df['total']=np.sum(df[['maths','physics','chemistry']],axis=1)
df['average']=df['total']/3
print(df['total'])
print(df['average'])
topper=df['total'].max()
print("topper:",topper)
topper_row = df[df['total'] == topper]
print(topper_row)
plt.bar(df['name'],df['total'],color='lightblue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.pie(df.loc[0, ['maths','physics','chemistry']], labels=['maths','physics','chemistry'], autopct='%1.1f%%')
plt.title(f"{df.loc[0, 'name']}'s Subject-wise Distribution")
plt.show()
```
